```python
import os
import pandas as pd
import numpy as np
import csv
```

```python
data_file = "purchase_data.csv"
purchase_df = pd.read_csv(data_file)
purchase_df
```

```python
players = len(purchase_df["SN"].unique())
total_players = pd.DataFrame({"Player Count":[players]})
total_players
```

```python
items = len(purchase_df["Item Name"].unique())
total_revenue = purchase_df["Price"].sum()
total_purchases = purchase_df["Purchase ID"].count()
average_purchase_price = round(purchase_df["Price"].mean(), 2)

purchase_analysis = pd.DataFrame({"Number of Unique Items": [items], "Average Purchase Price": [average_purchase_price], "Total Number of Purchase": [total_purchases] ,"Total Revenue": [total_revenue]})
purchase_analysis
```

```python
gender = purchase_df.groupby("Gender")["SN"].nunique()
gender
gender_count = gender.sum()
gender_count
male_percentage = round((gender[0]/gender_count)*100, 2)
male_count = gender[0]
female_percentage = round((gender[1]/gender_count)*100, 2)
female_count = gender[1]
unknown_gender_percentage = round((gender[2]/gender_count)*100, 2)
unknown_gender_count = gender[2]

gender_demographics =pd.DataFrame({"Male Count": [male_count], "Male Percentage":[f'{male_percentage}%'], "Female Count": [female_count], "Female Percentage":[f'{female_percentage}%'], "Gender Unknown": [unknown_gender_count], "Gender Unknown Percentage":[f'{unknown_gender_percentage}%']})
gender_demographics
```

```python
gender_purchases = purchase_df.groupby("Gender")["Item Name"].count()
gender_purchases

gender_purchases_average = purchase_df.groupby("Gender")["Price"].mean()
gender_purchases_average.head()

gender_purchases_total = purchase_df.groupby("Gender")["Price"].sum()
gender_purchases_total.head()

gender_average = gender_purchases_total/gender
gender_average

purchase_analysis_gender = pd.DataFrame({"Purchase Count": gender_purchases, "Average Purchase Price": round(gender_purchases_average,2), "Total Purchase Value": gender_purchases_total, "Average Purchase Total Per Person": round(gender_average,2)})
purchase_analysis_gender
```

```python
bins = [0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,99999]
group_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
purchase_df["Age Summary"] = pd.cut(purchase_df["Age"], bins, labels = group_names, include_lowest = True)
purchase_df
```

```python
agegroup_purchase = purchase_df.groupby("Age Summary")["SN"].count()
agegroup_purchase

agegroup_purchase_unique = purchase_df.groupby("Age Summary")["SN"].nunique()
agegroup_purchase_unique

agegroup_purchase_average = purchase_df.groupby("Age Summary")["Price"].mean()
agegroup_purchase_average

agegroup_purchase_total = purchase_df.groupby("Age Summary")["Price"].sum()
agegroup_purchase_total

agegroup_average = agegroup_purchase_total/agegroup_purchase_unique
agegroup_average

purchase_analysis_age = pd.DataFrame({"Purchase Count": agegroup_purchase, "Average Purchase Price": round(agegroup_purchase_average,2), "Total Purchase Value": agegroup_purchase_total, "Average Purchase Total Per Person": round(agegroup_average,2)})
purchase_analysis_age
```

```python
spending_total = purchase_df.groupby("SN")["Price"].sum()
spending_avg = purchase_df.groupby("SN")["Price"].mean()
spending_count = purchase_df.groupby("SN")["Price"].count()

top_spenders = pd.DataFrame({"Purchase Count": spending_count, "Average Purchase Price": spending_avg, "Total Purchase Value": spending_total})
top_spenders.sort_values("Total Purchase Value", ascending = False).head()
```

```python
popular_item = purchase_df[["Item ID", "Item Name", "Price"]]
item_details = popular_item.groupby(["Item ID", "Item Name"])
purchase_count_item = item_details["Price"].count()
purchase_value = (item_details["Price"].sum())
item_price = purchase_value/purchase_count_item

most_popular_items = pd.DataFrame({"Purchase Count": purchase_count_item, "Item Price": item_price, "Total Purchase Value": purchase_value})
most_popular_items.sort_values(["Purchase Count"], ascending =False).head()

```

```python
most_popular_items = pd.DataFrame({"Purchase Count": purchase_count_item, "Item Price": item_price, "Total Purchase Value": purchase_value})
most_popular_items.sort_values(["Total Purchase Value"], ascending =False).head()
```
